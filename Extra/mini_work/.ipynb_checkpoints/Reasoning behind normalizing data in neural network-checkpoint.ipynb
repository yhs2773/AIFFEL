Normalization, standardization, regularization: 셋 다 overfitting 막기 위해 사용
- Normalization: 학습 전에 공통 범위(0~1)로 바꿔주는 것. 흔히 min max scaling 사용
- Standardization: 학습 전에 데이터의 범위를 평균 0, 분산 1이 되게 바꿔주는 것. 흔히 z-score / iqr method 사용
- Regularization: weight를 조정하는데 규제를 검. L1 regularization, L2 regularization 등이 있음. 주로 하이퍼 파라미터를 조정하는 방식
    - 예: 학습 수를 짧게 조절, 특성을 골라서 학습, 앙상블 등등

DL에서는 rescaling 하는 과정이 중요한 이유:
- 학습을 더 빠르고 용이하게 해준다. 즉, 시간과 리소스를 아낄수 있다.
- 데이터가 서로 다른 범위(feature)로 인해서 한쪽으로 치우지지 않게 하기 위해

- 또 다른 문제인 plateau 문제도 해결하기 위해
