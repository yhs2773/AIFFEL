```python
import os
import requests
from io import BytesIO
from itertools import product
from datetime import datetime

# 데이터 처리
import pandas as pd
import numpy as np

# 시각화
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.ticker as plticker
import seaborn as sns

# 분석
import sklearn
import statsmodels.api as sm
import pmdarima as pm  
import statsmodels
from sklearn.metrics import *
from sklearn.ensemble import *
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge, RidgeCV, Lasso, LassoCV
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor

# statsmodels
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.seasonal import STL
from statsmodels.tsa.stattools import adfuller # 정상성 검정
from statsmodels.tsa.arima_model import ARIMA

import warnings
warnings.filterwarnings('ignore')
```

```python
file_path_raws = os.getenv('HOME') + '/aiffel/demand_forecasting/data/raws.csv'
df_raws = pd.read_csv(file_path_raws)

df_op_ts = df_raws.set_index('target_dates').sort_index()
df_op_ts.head()
```

# Step 1. 학습 데이터와 테스트 데이터 나누기

```python
# 설명 변수 테이블 설정
_x_data = df_raws[['click_d_1', 'click_d_2', 'click_d_3', 'click_d_4', 'click_d_5', 'click_d_6', 'click_d_7', 'is_clean', 'avg_precipitation', 'avg_temperature', 'target_dates']]

_x_weekday = pd.get_dummies(df_raws['weekday'], prefix='weekday', drop_first=False)

_x_data = pd.concat([_x_data, _x_weekday], axis=1)

_x_data.set_index("target_dates", drop=True, inplace=True)
_x_data = _x_data.sort_values('target_dates')

# 예측 대상 데이터 설정
target = df_raws[['op_rate_0d_all_cars', 'target_dates' ]]

target.set_index("target_dates", drop=True, inplace=True)
target = target.sort_values('target_dates')

# Train/Test 데이터셋 나누기
_x_data_train = _x_data[_x_data.index<'2021-04-01']
_x_data_test = _x_data[_x_data.index>='2021-04-01']
target_train = target[target.index<'2021-04-01']
target_test = target[target.index>='2021-04-01']
```

```python
x_train = _x_data_train
x_test = _x_data_test
y_train = target_train
y_test = target_test
```

# Step 2. Grid search를 이용하여 최상의 파라미터 조합 찾기

```python
# 모델별 rmse 저장
models_rmse = []
```

```python
# 모수
p = range(0, 2)
q = range(0, 2)
d = range(0, 2)
P = range(0, 2)
Q = range(0, 2)
D = range(0, 2)

# 초기화
model = sm.tsa.SARIMAX(endog=y_train, exog=x_train, order=(0, 0, 0), seasonal_order=(0, 0, 0, 7))
aic = float('inf')
result = None

# for 문으로 최소 AIC 찾기 (Grid Search)
for num, (
    tmp_p, tmp_d, tmp_q
    , tmp_P, tmp_Q, tmp_D
) in enumerate(list(product(p, q, d, P, Q, D)), 1):
    tmp_model = sm.tsa.SARIMAX(endog=y_train, exog=x_train
                               , order=(tmp_p, tmp_d, tmp_q)
                               , seasonal_order=(tmp_P, tmp_D, tmp_Q, 7))
    tmp_result = tmp_model.fit(disp=False) # or 0 to not show convergence information
    tmp_aic = tmp_result.aic
    print(f'iter: {num}, AIC: {tmp_aic}')
    if tmp_aic < aic:
        aic = tmp_aic
        model = tmp_model
        result = tmp_result
```

# Step3. 모형 구조 확인

```python
result.summary()
```

```python
result.plot_diagnostics(figsize=(10,10))
plt.show()
```

# Step 4. 차량 수요 예측

```python
# RMSE
fc = result.get_forecast(len(y_test), exog=x_test)
fc_pm = fc.predicted_mean
fc_ci = fc.conf_int()
sarimax_rmse = mean_squared_error(y_test, fc_pm, squared=False)
print('RMSE:', sarimax_rmse)
models_rmse.append(sarimax_rmse)
```

```python
# 시각화
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(12, 8))
plt.xticks(rotation=45)
loc = plticker.MultipleLocator(base=7.0) 
ax.xaxis.set_major_locator(loc)

plt.plot(y_train[-50:], alpha=0.5, color='black', label='training')
plt.plot(y_test, alpha=0.5, color='red', label='actual operation rate')
plt.plot(y_test.index.values, fc_pm, alpha=0.5, color='blue', label='predicted operation rate')
plt.fill_between(y_test.index.values, fc_ci.iloc[:, 0], fc_ci.iloc[:, 1], alpha=0.1, color='b') # Confidence Interval
plt.legend()
plt.show()
```
