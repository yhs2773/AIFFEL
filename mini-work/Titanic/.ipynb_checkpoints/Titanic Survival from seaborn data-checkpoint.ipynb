```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme(style='whitegrid')
```

```python
# check dataset names
sns.get_dataset_names()
```

```python
# load titanic data from seaborn
titanic = sns.load_dataset('titanic')
titanic
```

```python
# df info
titanic.info()
```

```python
# check the number of null data
titanic.isnull().sum()
```

```python
# check what's in the deck column
titanic['deck'].value_counts(dropna=False)
```

```python
# too many NaN, so removing the deck column
titanic = titanic.drop(columns='deck', axis=1)
```

```python
# fill NaN in age column with median values
titanic['age'] = titanic['age'].fillna(titanic['age'].median())
```

```python
# fill embarked and embark town column NaN values with mode values
titanic['embarked'].mode()
titanic['embark_town'].mode()
titanic['embarked'] = titanic['embarked'].fillna('S')
titanic['embark_town'] = titanic['embark_town'].fillna('Southampton')
```

```python
titanic.isnull().sum()
```

```python
# checking dupe vals
titanic[titanic.duplicated()]
```

```python
# new df of only survived
titanic_survived = titanic[titanic['survived'] == 1]
titanic_survived
```

```python
pie_data = titanic.groupby(['alive']).count().sex
plt.figure(figsize=(10,10))
plt.title('Percentage of survival from disaster', fontsize=12)
plt.pie(pie_data, labels=pie_data.index, autopct='%2.2f%%')
plt.show()
```

```python
# survival by pclass and sex
sns.displot(data=titanic_survived, x='sex', col='pclass')
```

```python
# survival by pclass and age
sns.displot(data=titanic_survived, x='age', col='pclass')
```
