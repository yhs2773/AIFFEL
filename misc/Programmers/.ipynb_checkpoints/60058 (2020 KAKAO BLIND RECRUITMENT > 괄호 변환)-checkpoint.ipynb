```python
# proto 1
def solution(p):
    answer = ''
    # 1. create variables
    cnt = 0
    u = ''
    v = ''
    # 2. run for loop for string p
    for i in range(len(p)):
        if p[i]== '(':
            cnt += 1
        else:
            cnt -= 1
        # split p into u and v, where u is complete and v is not
        if cnt < 0:
            u = p[:i]
            v = p[i:]
            break
    # 3. answer is p if no change is needed
    if cnt == 0:
        answer = p
    else:
        # 4. fixing v
        w = len(v)  # assigning original length of v to w
        for j in range(len(v)):
            if v[j] == '(' and v[j+1] == ')':
                v += '()'
                continue
            elif v[j] == '(':
                v += ')'
            elif v[j] == ')' and j != w-1: # if j is last of w, then don't change
                v += '('
        v = v[w:]           # slicing v to the original length of v
        answer = u + v      # answer will be u + v
    return answer
```

```python
# proto 2
def solution(p):
    answer = ''
    
    # create variables
    lcnt = 0
    rcnt = 0
    ucnt = 0
    u = ''
    v = ''
    
    # step 1. empty string, answer = empty string
    if p == '':
        return ''
    else:
        # step 2. split p into u and v when the cnts are equal
        for i in range(len(p)):
            if p[i] == '(':
                lcnt += 1         # count left parenthesis
            else:
                rcnt += 1         # count right parenthesis
                
            if lcnt == rcnt:      # if equal
                u = p[:i+1]       # split u upto i+1
                v = p[i+1:]       # split v from i+1
                break
        
        # step 3 if u has correct order of parenthesis, worry about v
        for j in range(len(u)):
            if u[j] == '(':
                ucnt += 1
            else:
                ucnt -= 1
            if ucnt < 0:    # this indicates of an incorrect order
                break        
        # if correct, head to 3-1
        if ucnt == 0:
            # step 3-1. recursive (v) back to step 1 until 3
            return u + solution(v)
        else:
            # step. 4-1 add '(' at the beginning 
            # step. 4-2 add recursive (v)
            # step. 4-3 add ')' at the end
            # step. 4-4 remove first and last elements, flip u, then add to w
            return '(' + solution(v) + ')' + u[-2:0:-1]
```

```python
# answer
def solution(p):
    answer = ''
    
    # create variables
    lcnt = 0
    rcnt = 0
    ucnt = 0
    u = ''
    v = ''
    
    # step 1. empty string, answer = empty string
    if p == '':
        return ''
    else:
        # step 2. split p into u and v when the cnts are equal
        for i in range(len(p)):
            if p[i] == '(':
                lcnt += 1         # count left parenthesis
            else:
                rcnt += 1         # count right parenthesis
                
            if lcnt == rcnt:      # if equal
                u = p[:i+1]       # split u upto i+1
                v = p[i+1:]       # split v from i+1
                break
        
        # step 3 if u has correct order of parenthesis, worry about v
        for j in range(len(u)):
            if u[j] == '(':
                ucnt += 1
            else:
                ucnt -= 1
            if ucnt < 0:    # this indicates of an incorrect order
                break        
        # if correct, head to 3-1
        if ucnt == 0:
            # step 3-1. recursive (v) back to step 1 until 3
            return u + solution(v)
        else:
            # step. 4-1 add '(' at the beginning 
            # step. 4-2 add recursive (v)
            # step. 4-3 add ')' at the end
            answer = '(' + solution(v) + ')'
            # step. 4-4 remove first and last elements, flip u, then add to w
            for i in u[1:-1]:
                if i == '(':
                    answer += ')'
                else:
                    answer += '('
            return answer
# the issue was me wanting to flip the whole u, whereas the problem wanted me to flip every single parenthesis in 'u'
# i.e.
# proto 2 : "))(()(" -> "()(())" (X)
# answer  : "))(()(" -> "(())()" (O)
```

```python
# most upvoted answer
def solution(p):
    if p=='':
        return p
    r=True
    c=0
    for i in range(len(p)):
        if p[i]=='(':
            c-=1
        else:
            c+=1
        if c>0:
            r=False
        if c==0:
            if r:
                return p[:i+1]+solution(p[i+1:])
            else:
                return '('+solution(p[i+1:])+')'+''.join(list(map(lambda x:'(' if x==')' else ')',p[1:i]) ))
```
